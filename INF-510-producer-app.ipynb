```python
import pandas as pd
import numpy as np
import io
import requests
```

```python
url="https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
s=requests.get(url).content
c=pd.read_csv(io.StringIO(s.decode('utf-8')),header=None)
c
```

```python
c.iloc[:,1] = c.iloc[:,1].map({ 'M' : 1, 'B' : 0})
c
```

```python
data = c.values
unique, counts = np.unique(data[:,1], return_counts=True)
stat = dict(zip(unique, counts))
num_0 = int(stat[0.0])
num_1 = int(stat[1.0])
stat,num_0,num_1
```

```python
num_test_1 = int(stat[1.0]*0.2)
num_test_0 = int(stat[0.0]*0.2)
num_test_1,num_test_0
```

```python
test = pd.concat([c[c.iloc[:,1] == 1].head(num_test_1), c[c.iloc[:,1] == 0].head(num_test_0)])
train = pd.concat([c[c.iloc[:,1] == 1].iloc[num_test_1:], c[c.iloc[:,1] == 0].iloc[num_test_0:]])
test_X = test.iloc[:,2:]
train_X = train.iloc[:,2:]
test_y = test.iloc[:,1]
train_y = train.iloc[:,1]
test_X = test_X.values
test_y = test_y.values
train_X = train_X.values
train_y = train_y.values
```

```python
from sklearn.preprocessing import normalize
from sklearn import svm
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,average_precision_score,roc_curve, auc, roc_auc_score, mean_squared_error,accuracy_score,recall_score,f1_score
```

```python
test_X_n = normalize(test_X)
train_X_n = normalize(train_X)
```

```python
accuracies_sl_train = []
accuracies_sl_test = []
precisions_sl_train = []
precisions_sl_test = []
recalls_sl_train = []
recalls_sl_test = []
F_scores_sl_train = []
F_scores_sl_test = []
auc_scores_sl_train = []
auc_scores_sl_test = []

import warnings
warnings.filterwarnings('ignore')
penalties = [1e-4,1e-3,1e-2,1e-1,1,1e1]
vals = []
for i in penalties:
    clf = svm.LinearSVC(C=i)
    val = cross_val_score(clf, train_X_n, y=train_y, cv=5)
    vals.append(np.mean(val))
p = penalties[np.argmax(vals)]
## fit with best p
clf = svm.LinearSVC(C=p)
clf = clf.fit(train_X_n, train_y)
y_pred_test = clf.predict(test_X_n)
y_pred_train = clf.predict(train_X_n)
confusion_matrix_train = confusion_matrix(train_y,y_pred_train)
confusion_matrix_test = confusion_matrix(test_y,y_pred_test)
fpr_test, tpr_test, _ = roc_curve(test_y, y_pred_test)
fpr_train, tpr_train, _ = roc_curve(train_y, y_pred_train)
auc_score_test = auc(fpr_test,tpr_test)
auc_score_train = auc(fpr_train,tpr_train)
accuracy_test = accuracy_score(test_y,y_pred_test)
accuracy_train = accuracy_score(train_y,y_pred_train)
precision_test = average_precision_score(test_y,y_pred_test)
precision_train = average_precision_score(train_y,y_pred_train)
recall_test = recall_score(test_y,y_pred_test)
recall_train = recall_score(train_y,y_pred_train)
F_score_test = f1_score(test_y,y_pred_test)
F_score_train = f1_score(train_y,y_pred_train)
print('Trainning confusion matrix\n',confusion_matrix_train,'\nTesting confusion matrix\n',confusion_matrix_test)
plt.plot(fpr_test, tpr_test, label='test')
plt.plot(fpr_train, tpr_train, label='train')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve')
plt.legend(loc='best')
plt.show()

accuracies_sl_train.append(accuracy_train)
accuracies_sl_test.append(accuracy_test)
precisions_sl_train.append(precision_train)
precisions_sl_test.append(precision_test)
recalls_sl_train.append(recall_train)
recalls_sl_test.append(recall_test)
F_scores_sl_train.append(F_score_train)
F_scores_sl_test.append(F_score_test)
auc_scores_sl_train.append(auc_score_train)
auc_scores_sl_test.append(auc_score_test)
```
